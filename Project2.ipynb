```python
import pandas as pd
import numpy as np
```

```python
content = pd.read_csv("./Junyi/Info_Content.csv")
user = pd.read_csv("./Junyi/Info_UserData.csv")
```

```python
chunks = []

for chunk in pd.read_csv("./Junyi/Log_Problem.csv", chunksize=1000000):
    chunks.append(chunk)

problems = pd.concat(chunks, axis=0)
```

```python
from deep_translator import GoogleTranslator
from concurrent.futures import ThreadPoolExecutor

def translate_text(text):
    return GoogleTranslator(source='auto', target='en').translate(text)

executor = ThreadPoolExecutor(max_workers=50)

content_sub = content.loc[1:500]

content['content_pretty_name'] = list(executor.map(translate_text, content['content_pretty_name']))
content
```

```python


# Assuming log_problem_df is your DataFrame containing the log problem data
unique_kids = pd.Series(problems['uuid'].unique())
sampled_kids = unique_kids.sample(n=100, random_state=1)

# Filter the log_problem_df DataFrame to only include rows corresponding to the sampled kids
sampled_log_problem_df = problems[problems['uuid'].isin(sampled_kids)]

# Perform the left join operation to combine the sampled log problem data with the exercise and content information
# Assuming info_content_df and info_userdata_df are your DataFrames containing the content and user data
merged_df = pd.merge(sampled_log_problem_df, content, on='ucid', how='left')
merged_df = pd.merge(merged_df, user, on='uuid', how='left')
```

```python
def to_6_digit_hex(i):
    return format(i, '06X')

# For uuid
uuid_mapping = {uuid: to_6_digit_hex(i+1) for i, uuid in enumerate(merged_df['uuid'].unique())}
merged_df['uuid'] = merged_df['uuid'].map(uuid_mapping)

# For ucid
ucid_mapping = {ucid: to_6_digit_hex(i+1) for i, ucid in enumerate(merged_df['ucid'].unique())}
merged_df['ucid'] = merged_df['ucid'].map(ucid_mapping)

# For upid
upid_mapping = {upid: to_6_digit_hex(i+1) for i, upid in enumerate(merged_df['upid'].unique())}
merged_df['upid'] = merged_df['upid'].map(upid_mapping)
```

```python
# Sort by 'uuid' and then by 'timestamp_TW'
merged_df.sort_values(by=['uuid', 'timestamp_TW'], inplace=True)
merged_df['timestamp_TW'] = pd.to_datetime(merged_df['timestamp_TW'])
merged_df['timestamp_TW'] = merged_df['timestamp_TW'].astype(int) / 10**9
```

```python
timestamp_counts = merged_df.groupby('uuid')['timestamp_TW'].count()

# Calculate quantiles
quantiles = timestamp_counts.quantile([0.25, 0.5, 0.75])

print("Quantiles of the number of timestamps for each user:")
print(quantiles)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.preprocessing.sequence import pad_sequences

# Assuming merged_df is your DataFrame containing the merged data
# Replace this with your actual DataFrame
# merged_df = ...

# Group the data by 'uuid'
grouped = merged_df.groupby('uuid')

X_train_sequences = []
y_train_sequences = []
X_test_sequences = []
y_test_sequences = []

# Define the sequence length
sequence_length = 20

# Get the index of the target variable 'is_correct'
target_index = merged_df.columns.get_loc('is_correct')

# Get the feature indices, excluding the target variable
feature_indices = [merged_df.columns.get_loc(col) for col in merged_df.columns if col != 'is_correct']

for name, group in grouped:
    if len(group.values) < 20:
        continue
    values = group.values
    
    train, test = train_test_split(values, test_size=0.2, shuffle=False)
    
    for i in range(len(train) - sequence_length):
        X_train_sequences.append(train[i:i + sequence_length, feature_indices])
        y_train_sequences.append(train[i + sequence_length, target_index])
        
    for i in range(len(test) - sequence_length):
        X_test_sequences.append(test[i:i + sequence_length, feature_indices])
        y_test_sequences.append(test[i + sequence_length, target_index])

# Convert lists to numpy arrays
X_train_sequences = np.array(X_train_sequences)
y_train_sequences = np.array(y_train_sequences)
X_test_sequences = np.array(X_test_sequences)
y_test_sequences = np.array(y_test_sequences)

# Pad sequences
#X_train_sequences = pad_sequences(X_train_sequences, maxlen=sequence_length, dtype='float32')
#X_test_sequences = pad_sequences(X_test_sequences, maxlen=sequence_length, dtype='float32')

# Reshape X_sequences to be 3D (samples, timesteps, features)
X_train_sequences = np.reshape(X_train_sequences, (X_train_sequences.shape[0], sequence_length, len(feature_indices)))
X_test_sequences = np.reshape(X_test_sequences, (X_test_sequences.shape[0], sequence_length, len(feature_indices)))
```

```python
# Build and compile the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(X_train_sequences.shape[1], X_train_sequences.shape[2])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Fit the model
model.fit(X_train_sequences, y_train_sequences, epochs=50, batch_size=32, verbose=1, validation_data=(X_test_sequences, y_test_sequences))
```
